## Honey Production

This proejct investigates the trend of of production of honey in U.S. and predicts the future production of honey based on the past observations.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model

# Load the data
df = pd.read_csv("honeyproduction.csv")

# Inspect the data
print(df.head())
print(df.columns)

# Calculate teh total production of honey per year
prod_per_year = df.groupby('year').totalprod.mean().reset_index()

# Create a variable called X that is the column of years in this prod_per_year DataFrame
X = prod_per_year['year'].values.reshape(-1, 1)
# print(X[:10])

# Create a variable called y that is the totalprod column in the prod_per_year dataset
y = prod_per_year['totalprod']
# print(y[:10])

# Plot y vs X as a scatterplot
plt.figure()
plt.scatter(X, y)
plt.show()

# Create a linear regression model
regr = linear_model.LinearRegression()

# Fit the model to the data 
regr.fit(X, y)

# Print out the slope and intercept of the line
print(regr.coef_)
print(regr.intercept_)

# Predict the y values on the X data
y_predict = regr.predict(X)

# Plot y_predict vs X
plt.figure()
plt.scatter(X, y)
plt.plot(X, y_predict)
plt.show()

# Create a numpy array ranges from 2013 to 2050
X_feature = np.array(range(2013, 2051))
X_future = X_feature.reshape(-1, 1)

# Create a list of future_predict based on X_freature
future_predict = regr.predict(X_future)

# Plot the future_predict vs X_future
plt.figure()
plt.scatter(X, y)
plt.plot(X_future, future_predict)
plt.show()
```

The production of honey has been declining from 1998 to 2012. We expect that the production of honey will be decling in the future.
